## Handling Missing Value in Machine Learning

### Impute missing values with Machhine Learning Prediction model

Youtube explanation : https://youtu.be/_QWLezcCKks

```python
import pandas as pd
import numpy as np
```

```python
# Import the dataset
df_titanic = pd.read_csv("https://raw.githubusercontent.com/atulpatelDS/Data_Files/master/Titanic/titanic_train.csv")
```

```python
df_titanic.info()
```

```python
df_titanic.head()
```

```python
df_titanic = df_titanic[["Survived", "Pclass", "Sex", "SibSp", "Parch", "Fare", "Age"]]
```

```python
df_titanic.head()
```

```python
## Convert "Sex" male=1 and female=0
df_titanic["Sex"].replace(["male","female"],[1,0],inplace=True)
```

```python
df_titanic.head()
```

```python
df_titanic.info()
```

```python
## Test Data -- Null value in "Age"
df_test = df_titanic[df_titanic["Age"].isnull()]
df_test
```

```python
df_test.shape
```

```python
X_test = df_test.drop("Age",axis=1)
X_test.head()
```

```python
## Train Dataset ("Age") column without null
df_train = df_titanic.dropna()
```

```python
df_train.shape
```

```python
df_train
```

```python
X_train = df_train.drop("Age",axis=1) ## all the features
X_train.head()
```

```python
X_train.shape
```

```python
y_train = df_train["Age"] # Label
y_train.head()
```

```python
y_train.shape
```

```python
print(X_test.shape)
print(type(X_test))
```

```python
## We will use Linear Regression ML model to predict the null value in "Age"
from sklearn.linear_model import LinearRegression
```

```python
## Fit the model
LR_model = LinearRegression()
LR_model.fit(X_train,y_train)
```

```python
## Lets Predict the y_test, It means "Age" values corresponding to those rows where "Age" has null values.
y_pred = LR_model.predict(X_test)
```

```python
print(y_pred.shape)
print(type(y_pred))
```

```python
X_test
```

```python
y_pred
```

```python
X_test["Age"] = y_pred
```

```python
X_test
```

```python
df_final = pd.concat([df_train,X_test])
```

```python
df_final.shape
```

```python
df_final
```

```python
df_final.sort_index(inplace=True)
```

```python
df_final
```

```python
df_final.info()
```

```python
df_final.Age = np.int64(df_final.Age)
```
